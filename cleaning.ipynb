```python
## here I am trying to check which use calendar days and which departments use business days
business_days = ['4 Business Days', '5 Business Days', '3 Business Days', '10 Business Days', '7 Business Days',
                 '2 Business Days', '1 Business Days', '14 Business Days', 'Business Days']

def is_business_day(value):
    return value in business_days
df311['is_business_days'] = df311['ERT (Estimated Response Time)'].apply(is_business_day)

df311.head()
```

```python
departments_with_calendar_days = df311[df311['is_business_days'] == False]['Department'].tolist()
departments_with_business_days = df311[df311['is_business_days'] == True]['Department'].tolist()

departments_with_calendar_days = (set(departments_with_calendar_days))
print(departments_with_calendar_days)
```

```python

departments_with_business_days = (set(departments_with_business_days))
print(departments_with_business_days)
```

```python
intersection = departments_with_business_days.intersection(departments_with_calendar_days)
intersection
```

```python
busines_days_only = departments_with_business_days - departments_with_calendar_days
calendar_days_only = departments_with_calendar_days - departments_with_business_days

print(busines_days_only)
print()
print(calendar_days_only)
```

So meaningful insight found is that no department only uses calendar days but some uses business days and do not use calendar days and some use both
